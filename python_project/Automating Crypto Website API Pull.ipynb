```python
from requests import Request, Session
from requests.exceptions import ConnectionError, Timeout, TooManyRedirects
import json

url = 'https://pro-api.coinmarketcap.com/v1/cryptocurrency/listings/latest'
parameters = {
  'start':'1',
  'limit':'15',
  'convert':'USD'
}
headers = {
  'Accepts': 'application/json',
  'X-CMC_PRO_API_KEY': 'c24efd3c-f038-45f8-897e-402ad1658d99',
}

session = Session()
session.headers.update(headers)

try:
  response = session.get(url, params=parameters)
  data = json.loads(response.text)
  print(data)
except (ConnectionError, Timeout, TooManyRedirects) as e:
  print(e)
    
```

```python
import pandas as pd
pd.set_option("display.max_columns", None)
```

```python
df3 = df.groupby("name", sort=False)[["quote.USD.percent_change_1h", "quote.USD.percent_change_24h", "quote.USD.percent_change_7d", "quote.USD.percent_change_30d", "quote.USD.percent_change_60d", "quote.USD.percent_change_90d"]].mean(numeric_only=True)
df3.transpose()
```

```python
df4 = df3.stack()
df4
```

```python
df5 = df4.to_frame(name="values")
df5
```

```python
df5 = df5.reset_index()
df5 = df5.rename(columns={"level_1":"percent_change"})
df5
```

```python
df5["percent_change"] = df5["percent_change"].replace(["quote.USD.percent_change_1h","quote.USD.percent_change_24h", "quote.USD.percent_change_7d","quote.USD.percent_change_30d","quote.USD.percent_change_60d","quote.USD.percent_change_90d"],
                                                      ["1h","24h","7d","30d","60d","90d"])
df5
```
